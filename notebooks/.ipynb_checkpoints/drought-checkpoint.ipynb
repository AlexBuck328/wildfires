```python
# Load desired packages

%matplotlib inline

import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import fiona

# change default figsize
plt.rcParams['figure.figsize'] = (55, 35)

# plot style
plt.style.use('seaborn')
```

```python
droughtRaw = gpd.read_file('../rawData/droughtShape/drought.shp')
```

```python
droughtRaw
```

```python
droughtRaw.plot();
```

```python
droughtRawZero = droughtRaw[(droughtRaw.DM == 0)]

droughtRawOne = droughtRaw[(droughtRaw.DM == 1)]

droughtRawTwo = droughtRaw[(droughtRaw.DM == 2)]

droughtRawThree = droughtRaw[(droughtRaw.DM == 3)]
```

```python
fig, ax = plt.subplots()

droughtRawZero.plot(ax=ax, color='white', zorder=0)
droughtRawOne.plot(ax=ax, color='yellow', zorder=1)
droughtRawTwo.plot(ax=ax, color='orange', zorder=2)
droughtRawThree.plot(ax=ax, color='red', zorder=3)

ax.set(title="Drought",
       xlabel="X coordinates in degrees",
       ylabel="Y coordinates in degrees");
```
