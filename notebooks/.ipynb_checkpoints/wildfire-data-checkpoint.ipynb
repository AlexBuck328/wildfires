# Cleaning Data

Removing null records and dropping unwanted columns

```python
# Drop unwanted columns

firePoints.drop(['OBJECTID', 'FOD_ID', 'FPA_ID', 'SOURCE_SYSTEM_TYPE', 'SOURCE_SYSTEM', 'NWCG_REPORTING_AGENCY', 
               'NWCG_REPORTING_UNIT_ID', 'NWCG_REPORTING_UNIT_NAME', 'SOURCE_REPORTING_UNIT', 
               'SOURCE_REPORTING_UNIT_NAME', 'LOCAL_FIRE_REPORT_ID', 'LOCAL_INCIDENT_ID', 
               'FIRE_CODE', 'ICS_209_INCIDENT_NUMBER', 'ICS_209_NAME', 'MTBS_ID', 'MTBS_FIRE_NAME', 'STAT_CAUSE_CODE', 
               'FIRE_SIZE_CLASS'], axis=1, inplace=True)
```

```python
# Filter for records with Discovery and Contained Dates

firePointsNoNulls = firePoints[(firePoints.CONT_DATE.notnull()) & (firePoints.DISCOVERY_DATE.notnull())]

firePointsNoNulls
```

```python
fpY2K = firePointsNoNulls[firePointsNoNulls.FIRE_YEAR >= 2000]

fpY2K
```

```python
fpY2K.to_file('../rawData/pointsY2K.geojson', driver='GeoJSON')
```

```python
# Create a file for data with Ignition and Control dates

firePointsNoNulls.to_file('../rawData/pointsNoNulls.geojson', driver='GeoJSON')
```

```python
# Create subset without null Contained Dates and null FIPS Codes

noNulls = firePoints[(firePoints.CONT_DATE.notnull()) & (firePoints.FIPS_CODE.notnull())]
```

```python
# Verify Contained Date and FIPS Code columns do not have null values

noNulls.isnull().sum()
```

```python
# Take a look at the data

# show all columns
pd.set_option('display.max_columns', None)

noNulls
```

```python
# Drop unwanted columns

noNulls.drop(['OBJECTID', 'FOD_ID', 'FPA_ID', 'SOURCE_SYSTEM_TYPE', 'SOURCE_SYSTEM', 'NWCG_REPORTING_AGENCY', 
               'NWCG_REPORTING_UNIT_ID', 'NWCG_REPORTING_UNIT_NAME', 'SOURCE_REPORTING_UNIT', 
               'SOURCE_REPORTING_UNIT_NAME', 'LOCAL_FIRE_REPORT_ID', 'LOCAL_INCIDENT_ID', 
               'FIRE_CODE', 'ICS_209_INCIDENT_NUMBER', 'ICS_209_NAME', 'MTBS_ID', 'MTBS_FIRE_NAME', 'STAT_CAUSE_CODE', 
               'FIRE_SIZE_CLASS'], axis=1, inplace=True)
```

```python
# Create a file for data with Ignition and Control dates

noNulls.to_file('../rawData/pointsNoNulls.geojson', driver='GeoJSON')
```

```python
# Load the newly created data set

pointsNoNulls = gpd.read_file('../rawData/pointsNoNulls.geojson')
```

```python
# Verify data loaded

pointsNoNulls
```

```python
# Convert Discovery Date and Cont Date (contained date) columns to datetime

pointsNoNulls['DISCOVERY_DATE'] = pd.to_datetime(pointsNoNulls.DISCOVERY_DATE)

pointsNoNulls['CONT_DATE'] = pd.to_datetime(pointsNoNulls.CONT_DATE)
```

```python
# See how datetime conversion changed Discovery Date and Contained Date columns

pointsNoNulls.DISCOVERY_DATE, pointsNoNulls.CONT_DATE
```

# Visualizing Data

Visually exploring our data. Doing so shows us trends that might have gone unnoticed while the data was in a tabular form. For instance, the first line graph shows that average fire size has increased over time. The second line graph shows that wildfires predominantly occur between May and September.

### Line Graphs:
x = year and y = acres

x = month and y = acres

### Bar Graphs:
x = state and y = acres

### Box Plot:
counties within states

```python
# Average fire size each year

acYearMean = pointsNoNulls.groupby('FIRE_YEAR').FIRE_SIZE.mean()

acYearMean.plot();
```

```python
# Average fire size each month

discMonth = pointsNoNulls.DISCOVERY_DATE.dt.month

acYearMean = pointsNoNulls.groupby(discMonth).FIRE_SIZE.mean()

acYearMean.plot();
```
