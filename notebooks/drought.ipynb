```python
# Load desired packages

%matplotlib inline

import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import fiona

# change default figsize
plt.rcParams['figure.figsize'] = (25, 15)

# plot style
plt.style.use('seaborn')
```

```python
states = gpd.read_file('../rawData/cb_2018_us_state_20m/cb_2018_us_state_20m.shp')
```

```python
contigStates = gpd.read_file('../rawData/States_shapefile-shp/States_shapefile.shp')
```

```python
droughtRaw = gpd.read_file('../rawData/droughtShape/drought.shp')
```

```python
droughtRaw
```

```python
droughtRaw.plot();
```

```python
droughtRawZero = droughtRaw[(droughtRaw.DM == 0)]

droughtRawOne = droughtRaw[(droughtRaw.DM == 1)]

droughtRawTwo = droughtRaw[(droughtRaw.DM == 2)]

droughtRawThree = droughtRaw[(droughtRaw.DM == 3)]
```

```python
fig, ax = plt.subplots(figsize=(25,25))

contigStates.plot(ax=ax, color='#2171b5', zorder=0)
droughtRawZero.plot(ax=ax, color='#fef0d9', zorder=1)
droughtRawOne.plot(ax=ax, color='#fdcc8a', zorder=2)
droughtRawTwo.plot(ax=ax, color='#fc8d59', zorder=3)
droughtRawThree.plot(ax=ax, color='#d7301f', zorder=4)

ax.set(title="Drought",
       xlabel="X coordinates in degrees",
       ylabel="Y coordinates in degrees");
```
